```python
import pandas as pd
import matplotlib.pyplot as plt

xlstm_logging_file = "logs_training_xlstm.csv"
gpt_logging_file = "logs_training_gpt.csv"

xlstm_data = pd.read_csv(xlstm_logging_file)[:4000]
gpt_data = pd.read_csv(gpt_logging_file)

# plt.plot(xlstm_data["Step"], smoothed_loss, label="Smoothed Loss", color="pink")
plt.plot(xlstm_data["Step"], xlstm_data["Loss"], alpha=1, label="xLSTM Loss", color="beige")
plt.plot(gpt_data["Step"], gpt_data["Loss"], alpha=1, label="GPT-2 Loss", color="brown")

plt.axhline(y=6, color='r', linestyle='--', alpha=0.6)
plt.axhline(y=5, color='gray', linestyle='--', alpha=0.6)
plt.axhline(y=4, color='y', linestyle='--', alpha=0.6)
plt.axhline(y=3, color='g', linestyle='--', alpha=0.6)

plt.xlabel("Step")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
